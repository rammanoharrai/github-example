```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup as BS
import folium
import json
from geopy.geocoders import Nominatim
from sklearn.cluster import KMeans
from pandas.io.json import json_normalize
import matplotlib.cm as cm
import matplotlib.colors as colors
```

Scrapping website mapsofindia for neighborhoods in bangalore

```python
bangalore_sites=requests.get('https://www.mapsofindia.com/pincode/india/karnataka/bangalore/').text
```

Extracting out locational information after looking at structure of html file

```python
bangalore_data=BS(bangalore_sites,'html.parser')
bangalore_table=bangalore_data.find('table').find_all('td')
```

Creating the list of releavant data

```python
j=[]
for l in bangalore_table:
    j.append(l.text.splitlines()[0])
```

Extracting the relevant data in separate lists

```python
pincode=j[1::8]
location=j[0::8]
state=j[2::8]
district=j[3::8]
```

Converting List into Dataframe

removing the first element as it contains header information in original html

```python
bangalore_df_clean=bangalore_df[1::]
```

```python
fixed_columns = [venues_south_east_bangalore_onehot.columns[-3]] +[venues_south_east_bangalore_onehot.columns[-2]]+[venues_south_east_bangalore_onehot.columns[-1]]+ list(venues_south_east_bangalore_onehot.columns[:-3])
venues_south_east_bangalore_onehot=venues_south_east_bangalore_onehot[fixed_columns]
venues_south_east_bangalore_onehot
```

```python
south_east_bangalore_category_prem=venues_south_east_bangalore_onehot.groupby('location').mean().reset_index()
```

```python
south_east_bangalore_category1=south_east_bangalore_category_prem[south_east_bangalore_category_prem['Indian Restaurant']!=1]
south_east_bangalore_category=south_east_bangalore_category1[south_east_bangalore_category1['Bakery']!=1]
```

Lets cluster them into 4 clusters. Clustering them in 5 didn't give better result

```python
print(south_east_bangalore_category_prem.shape)
south_east_bangalore_category
```

```python
kclusters = 4

south_east_bangalore_category_grouped_clustering = south_east_bangalore_category.drop(labels='location',axis=1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(south_east_bangalore_category_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
south_east_bangalore_category['cluster']=kmeans.labels_
```

Exploring clusters on Banglaore maps

```python

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]
```

```python

map_bangalore1 = folium.Map(location=[(mid_lat+bangalore_df_cleaned['latitude'].min())/2, (mid_lon+bangalore_df_cleaned['longitude'].max())/2], zoom_start=11)
for lat, lon, poi, cluster1 in zip(south_east_bangalore_category['latitude'], south_east_bangalore_category['longitude'], south_east_bangalore_category['location'], south_east_bangalore_category['cluster']):
    cluster=int(cluster1)
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_bangalore1)
```
